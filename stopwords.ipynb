```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
en_stopwords = stopwords.words('english')
print(en_stopwords)
```

```python
sentence= "the shop is very far .we have to take cab for it"
nostopword = ' '.join ([word for word in sentence.split() if word not in en_stopwords])
print(nostopword)
```
